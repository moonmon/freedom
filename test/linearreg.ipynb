```python
from sklearn.cross_validation import train_test_split
```

```python
dataframe_dr = dataframe.dropna().sort_values('Date')
dataframe_dr
```

```python
from datetime import date
all_Data = []
```

```python
#从开始的日期找出相差得天数
days_since_start = [(x - dataframe_dr['Date'].min()).days for x in dataframe_dr['Date']]
days_since_start
```

```python
dataframe_dr['Days'] = days_since_start
```

```python
#进行one-hot编码
suberb_dummuse = pd.get_dummies(dataframe_dr[['Type', 'Method']])
suberb_dummuse
```

```python
dataframe_dr.head()
```

```python
all_date = dataframe_dr.drop(["Address","Price","Date", "SellerG","Suburb","Type","Method","CouncilArea","Regionname"], axis=1).join(
suberb_dummuse)
all_date
```

```python
X = all_date
y = dataframe_dr['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
lm.intercept_
```

```python
X.columns
```

```python
#回归系数
coeff_df = pd.DataFrame(lm.coef_, X.columns, columns=['Coefficient'])
coeff_df
```

```python
ranked_suburbs  = coeff_df.sort_values('Coefficient', ascending=False)
ranked_suburbs
```

```python
predictions = lm.predict(X_test)
predictions
```

```python
plt.scatter(y_test, predictions)
plt.ylim([200000,1000000])
plt.xlim([200000,1000000])
```

```python
sns.distplot((y_test-predictions),bins=50)
```

```python
from sklearn import metrics
```

```python
#
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
